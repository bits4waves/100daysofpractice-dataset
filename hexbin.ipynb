```python
posts_per_day = pd.DataFrame(posts.groupby(posts.post_date).size())
```

```python
posts_per_day
```

```python
posts_per_day['post_date'] = posts_per_day.index
```

```python
posts_per_day
```

```python
[(d, pd.Timestamp(d)) for d in posts_per_day['post_date']]
```

```python
posts_per_day['date'] = [pd.Timestamp(d) for d in posts_per_day['post_date']]
```

```python
posts_per_day
```

```python
violincase_per_day = pd.DataFrame(violincase.groupby(violincase.date).size())
```

```python
violincase_per_day
```

```python
violincase_per_day['date'] = [pd.Timestamp(d) for d in violincase_per_day.index]
```

```python
violincase_per_day
```

```python
posts_per_day['violincase_posted'] = posts_per_day['date'].isin(violincase_per_day['date'])
```

```python
posts_per_day.tail(30)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates


df = posts_per_day

fig, ax = plt.subplots()

df.plot(ax=ax,
        title='Posts per day (all users)',
        x='date',
        xlim=(pd.to_datetime('2017-05-01'),
              pd.to_datetime('2021-04-29')),
        figsize=(12, 8))

for i, row in df.iterrows():
    if row['violincase_posted']:
        ax.axvline(x=row['date'],color='red',alpha=0.15)

ax.xaxis.set_major_locator(mdates.MonthLocator())

plt.setp(ax.get_xticklabels(), rotation=45, fontsize=9)

plt.show()
```
